```python
import pandas as pd
import random
from sklearn.metrics import classification_report

# Set random seed for reproducibility
random.seed(42)


df = pd.read_json("../../datasets/processed/claims/factool_qa/data_gpt-4o_annotated.json")

# Fix the label column
df["label"] = df["label"].apply(lambda x: True if x == "true" else False)

# Create a columns of random labels (True/False)
df["random_label"] = [
    random.choice([True, False]) for _ in range(len(df))
]

# Create a columns of All true labels
df["all_true_labels"] = True

# Create a columns of All false labels
df["all_false_labels"] = False

# Get the classification report for the random labels
print("Classification report for random labels:")
print(
    classification_report(
        df["random_label"],
        df["label"],
        zero_division=0,
    )
)

# Get the classification report for the all true labels
print("Classification report for all true labels:")
print(
    classification_report(
        df["all_true_labels"],
        df["label"],
        zero_division=0,
    )
)

# Get the classification report for the all false labels
print("Classification report for all false labels:")
print(
    classification_report(
        df["all_false_labels"],
        df["label"],
        target_names=["False", "True"],
        zero_division=0,
    )
)
```
